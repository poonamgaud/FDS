```python
import matplotlib.pyplot as plt
plt.figure(figsize=[15,5])
countries=['usa','brazil','russia','sapin','uk','india']
totalCase=[2343221,4324567,7685432,1243654,1456783,5643210]
totalDeath=[1154323,2344531,1123546,4566743,3487654,2311567]
for i in range (len(countries)):
    plt.bar(countries[i],totalDeath[i],bottom=totalCase[i]-totalDeath[i],color='blue')
    plt.bar(countries[i],totalCase[i]-totalDeath[i],color='orange')
plt.legend(labels=['total Deaths','total Cases'])
plt.title('bar plot representing the total deaths and total cases countries and total cases country wise')
plt.xlabel('countries')
plt.ylabel('Cases')
plt.savefig('3barplot.png')
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.figure(figsize=[15,10])
country=['usa','brazil','russia','sapin','uk','india']
totalDeath=[2343221,4324567,7685432,1243654,1456783,5643210]
totalReecovery=[1154323,2344531,1123546,4566743,3487654,2311567]
activeCases=[234522,2456561,234765,5432674,3455632,4907651]
x=np.arange(len(totalDeath))
plt.bar(x,totalDeath,color='blue',width=0.25)
plt.bar(x+0.25,totalReecovery,color='g',width=0.25)
plt.bar(x+0.5,activeCases,color='r',width=0.25)
plt.legend(['TotalDeath','TotalReecovery','ActiveCases'])
plt.xticks([i+0.25 for i in range(6)],country)
plt.title("bar plot representing the total deaths,total recovery cases and active cases wise")
plt.xlabel('countries')
plt.ylabel('cases')
plt.savefig('4barplot.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as lt
%matplotlib inline
labels=['Civil','Electrical','Mechanical','Chemical']
sizes=[15,50,47,10]
explode=(0.1,0.1,0.1,0.4)
fig,ax=plt.subplots()
ax.pie(sizes,
      explode=explode,
      labels=labels,
      autopct='%1.1f%%',
      shadow=True,
      startangle=90)
ax.axis('equal')
ax.set_title('Engineering Diciplines')
plt.show()
```
